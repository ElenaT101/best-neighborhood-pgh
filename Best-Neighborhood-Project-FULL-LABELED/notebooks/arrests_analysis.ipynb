# Arrest Rate Analysis

We use total arrest counts per neighborhood as a direct measure of criminal activity.

This metric is normalized and contributes equally to the final safety score.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../data/arrest_summary.csv')
top10 = df.sort_values('Arrest_Count').head(10)

plt.figure(figsize=(10, 6))
plt.bar(top10['Neighborhood'], top10['Arrest_Count'], color='lightblue')
plt.title('Top 10 Safest Neighborhoods by Arrests')
plt.xlabel('Neighborhood')
plt.ylabel('Arrest Count')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
# Highlight Stanton Heights' arrest count and compare to top 10 safest
highlight_arrest = df[df['Neighborhood'] == 'Stanton Heights']
top10_arrests = df.sort_values('Arrest_Count').head(10)

# Show how Stanton Heights compares
print("🚓 Stanton Heights - Total Arrests:", int(highlight_arrest['Arrest_Count'].values[0]))
print("\nTop 10 Safest Neighborhoods by Arrest Count:")
print(top10_arrests[['Neighborhood', 'Arrest_Count']])
```
