# Home Value Analysis

We use home value estimates as a proxy for housing stability and neighborhood investment. A higher total home count suggests a more established and stable community.

This metric is normalized and contributes equally to the final safety score.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../data/home_value_summary.csv')
top10 = df.sort_values('Total_Homes', ascending=False).head(10)

plt.figure(figsize=(10, 6))
plt.bar(top10['Neighborhood'], top10['Total_Homes'], color='orange')
plt.title('Top 10 Neighborhoods by Total Homes')
plt.xlabel('Neighborhood')
plt.ylabel('Estimated Total Homes')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Highlight Stanton Heights' position in home values
highlight = df[df['Neighborhood'] == 'Stanton Heights']
top10_compare = df.sort_values('Total_Homes', ascending=False).head(10)

# Show how Stanton Heights compares
print("🏡 Stanton Heights - Estimated Total Homes:", int(highlight['Total_Homes'].values[0]))
print("\nTop 10 Neighborhoods by Home Values:")
print(top10_compare[['Neighborhood', 'Total_Homes']])
```
