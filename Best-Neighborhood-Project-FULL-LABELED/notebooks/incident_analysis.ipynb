# Fire Incident Analysis

Fire incident reports offer a broader look at public safety. Fewer incidents signal a safer, quieter neighborhood.

This metric is normalized and contributes equally to the final safety score.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../data/incident_summary.csv')
top10 = df.sort_values('Incident_Count').head(10)

plt.figure(figsize=(10, 6))
plt.bar(top10['Neighborhood'], top10['Incident_Count'], color='red')
plt.title('Top 10 Safest Neighborhoods by Fire Incidents')
plt.xlabel('Neighborhood')
plt.ylabel('Incident Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Highlight Stanton Heights' incident count and compare to top 10 safest
highlight_incident = df[df['Neighborhood'] == 'Stanton Heights']
top10_incidents = df.sort_values('Incident_Count').head(10)

# Show how Stanton Heights compares
print(" Stanton Heights - Total Fire Incidents:", int(highlight_incident['Incident_Count'].values[0]))
print("\nTop 10 Safest Neighborhoods by Incident Count:")
print(top10_incidents[['Neighborhood', 'Incident_Count']])

```
